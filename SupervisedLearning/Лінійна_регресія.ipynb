```python
import numpy as np
import pandas as pd

data = {
    "Days": [list(map(int, input("Days: ").split()))],
    "Coins":[list(map(int, input("Coins: ").split()))]
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Days"], df["Coins"], color="blue")
plt.title("Графік залежності")
plt.xlabel("Days")
plt.ylabel("Coins")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Days"]]
Y = df["Coins"]

model = LinearRegression()
model.fit(X, Y)

print("Вага (W):", model.coef_)
print("Зміщення (b):", model.intercept_)
```

```python
Y_pred = model.predict(X)

plt.scatter(X, Y, color="blue")
plt.plot(X, Y_pred, color="red")
plt.title("Лінійна регресія")
plt.xlabel("Days")
plt.ylabel("Coins")
```

```python
new_value = [[10]]
predicted_coins = model.predict(new_value)
print(f"Прозноз на {new_value[0][0]} день: {predicted_coins[0]} монеток")
```
