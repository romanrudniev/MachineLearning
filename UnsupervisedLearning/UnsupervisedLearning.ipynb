```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
X = np.vstack((np.random.randn(50, 2) + [2, 2], np.random.randn(50, 2) + [7, 7]))
```

```python
plt.scatter(X[:,0], X[:,1], s=50, color='gray')
plt.title("Вихідні дані")
plt.xlabel('X1')
plt.ylabel('X2')
plt.show
```

```python
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels)
print(centroids)
```

```python
plt.scatter(X[:,0], X[:,1], s=50, c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Центроїди')
plt.title("Результат")
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show
```
