# Обробка жанрів

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
url = "movies.csv"

df = pd.read_csv(url)

df.head()
```

```python
def extract_genres(genres_str):
  try:
    genres = ast.literal_eval(genres_str)
    return [genre['name'] for genre in genres_str]
  except ValueError:
    return []
```

```python
df['genres']
```

```python
try:
  df['genres'] = df['genres'].apply(extract_genres)
except:
  df['genres'] = []


```

```python
genre_counts = df['genres'].value_counts()
print(genre_counts)
```
